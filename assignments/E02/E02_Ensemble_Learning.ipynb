# Questão 1

Implemente uma função load_data(seed) que:

Carregue o dataset `Fashion MNIST`
Realize a separação em treino e teste
Utilize `train_test_split` com controle de aleatoriedade
Retorne: `X_train`, `X_test`, `y_train`, `y_test`

Depois responda: 
É necessário normalizar os dados para esse tipo de modelo? Justifique.

**Solução**:

````python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score

def load_data(seed=42):
    data = fetch_openml('mnist_784', version=1, as_frame=False)
    X, y = data['data'], data['target']
    y = y.astype(int)
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=seed, stratify=y
    )
    return X_train, X_test, y_train, y_test
````

# Questão 2

Implemente as funções:

`train_random_forest(X_train, y_train, seed)`
`train_adaboost(X_train, y_train, seed)`

## Requisitos:

Utilizar os modelos do `sklearn`
Garantir reprodutibilidade com `random_state`

**Solução**:

````python

def train_random_forest(X_train, y_train, seed=42):
    clf = RandomForestClassifier(random_state=seed, n_estimators=100)
    clf.fit(X_train, y_train)
    return clf

def train_adaboost(X_train, y_train, seed=42):
    clf = AdaBoostClassifier(random_state=seed, n_estimators=50)
    clf.fit(X_train, y_train)
    return clf
````

# Questão 3

Implemente a função:

- `evaluate(model, X_test, y_test)`

Ela deve:
- Realizar predições
- Retornar a acurácia do modelo

**Solução**:

````python

def evaluate(model, X_test, y_test):
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    return acc
````

# Questão 4

Implemente a função:

- `run_pipeline(model_type="rf", seed=42)`

Ela deve:
- Carregar os dados
- Treinar o modelo escolhido (`rf` ou `ab`)
- Avaliar o modelo
- Retornar a acurácia

**Solução**:

````python
def run_pipeline(model_type="rf", seed=42):
    X_train, X_test, y_train, y_test = load_data(seed)

    if model_type == "rf":
        model = train_random_forest(X_train, y_train, seed)
    elif model_type == "ab":
        model = train_adaboost(X_train, y_train, seed)
    else:
        raise ValueError("Tipo de modelo inválido. Use 'rf' ou 'ab'.")
    
    acc = evaluate(model, X_test, y_test)
    
    return acc
````

# Questão 5

Execute o pipeline para ambos os modelos:

- Random Forest
- AdaBoost

## Apresente:
- Acurácia, Precisão, Recall e F1-Score de cada modelo

## Responda:
- Qual modelo apresentou melhor desempenho inicial?

**Solução:**
Para comparar os modelos, vamos calcular acurácia, precisão, recall e F1-Score para Random Forest e AdaBoost. O modelo com melhor desempenho inicial é aquele que apresentar os melhores valores nessas métricas.

````python
from sklearn.metrics import precision_score, recall_score, f1_score
X_train, X_test, y_train, y_test = load_data(42)


rf = train_random_forest(X_train, y_train, 42)
ab = train_adaboost(X_train, y_train, 42)
y_pred_rf = rf.predict(X_test)
y_pred_ab = ab.predict(X_test)

print('Random Forest:')
print('Acurácia:', accuracy_score(y_test, y_pred_rf))
print('Precisão:', precision_score(y_test, y_pred_rf, average='macro'))
print('Recall:', recall_score(y_test, y_pred_rf, average='macro'))
print('F1-Score:', f1_score(y_test, y_pred_rf, average='macro'))
print('---')
print('AdaBoost:')
print('Acurácia:', accuracy_score(y_test, y_pred_ab))
print('Precisão:', precision_score(y_test, y_pred_ab, average='macro'))
print('Recall:', recall_score(y_test, y_pred_ab, average='macro'))
print('F1-Score:', f1_score(y_test, y_pred_ab, average='macro'))
````

**Solução:**

Para pelo menos um dos modelos:

- Compare a acurácia em treino e teste

## Responda:
- Existe overfitting?
- Qual modelo tende a sofrer mais com isso?

**Solução:**
Para verificar overfitting, compare a acurácia no treino e no teste. Se a acurácia no treino for muito maior que no teste, há overfitting. Modelos mais complexos, como Random Forest, tendem a sofrer mais com isso se não forem bem regulados.

````python
X_train, X_test, y_train, y_test = load_data(42)
rf = train_random_forest(X_train, y_train, 42)

acc_train = rf.score(X_train, y_train)
acc_test = rf.score(X_test, y_test)

print('Acurácia treino:', acc_train)
print('Acurácia teste:', acc_test)
````

# Questão 8

Varie pelo menos um hiperparâmetro em cada modelo:

- Random Forest: `n_estimators`
- AdaBoost: `n_estimators`

## Analise:
- O desempenho muda significativamente?

## Responda:
- Qual modelo é mais sensível a mudanças?

**Solução:**
Ao variar o n_estimators, o desempenho dos modelos pode mudar. Random Forest tende a ser mais estável, enquanto AdaBoost pode ser mais sensível a esse hiperparâmetro.

````python
for n in [10, 50, 100, 200]:
    rf = RandomForestClassifier(n_estimators=n, random_state=42)
    ab = AdaBoostClassifier(n_estimators=n, random_state=42)

    X_train, X_test, y_train, y_test = load_data(42)

    rf.fit(X_train, y_train)
    ab.fit(X_train, y_train)
    
    print(f'Random Forest n={n}:', rf.score(X_test, y_test))
    print(f'AdaBoost n={n}:', ab.score(X_test, y_test))
````
